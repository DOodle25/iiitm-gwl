```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
fetal=pd.read_csv("C:/Users/dell/Documents/GitHub/iiitm-gwl/ML-Project/Dataset/fetal_health.csv")
```

```python
fetal.head()
```

```python
print(fetal.shape)
```

```python
model1=DecisionTreeClassifier(criterion='gini',random_state=42)
model2=DecisionTreeClassifier(criterion='entropy',random_state=42)
```

```python
y=fetal[['fetal_health']]
x=fetal.loc[:,fetal.columns!='fetal_health']
```

```python
y.head()
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.7,random_state=42)
```

```python
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train,y_train)
```

```python
model1.predict(x_test)
```

```python
model2.predict(x_test)
```
