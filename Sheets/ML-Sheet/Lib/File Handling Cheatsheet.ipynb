# File Handling Cheatsheet

Examples for reading and writing various file formats using pandas.

```python
# Reading CSV file
import pandas as pd
df_csv = pd.read_csv('./archive/iris.csv')
print(df_csv.head())
```

```python
# Reading Excel file (.xlsx)
df_excel = pd.read_excel('./converted/output.xlsx', engine='openpyxl')
print(df_excel.head())
```

```python
# Reading Excel file (.xls)
# df_xls = pd.read_excel('./converted/output.xls', engine='xlrd')
# print(df_xls.head())
```

```python
# Reading JSON file
df_json = pd.read_json('./converted/output.json')
print(df_json.head())
```

```python
# Writing DataFrame to CSV
df_csv.to_csv('./converted/output.csv', index=False)
```

```python
# Writing DataFrame to JSON
df_csv.to_json('./converted/output.json', orient='records')
```

```python
# Writing DataFrame to Excel
df_csv.to_excel('./converted/output.xlsx', index=False, engine='openpyxl')
```
