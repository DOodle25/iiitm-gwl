```python
import numpy as np
```

```python
y_true = np.array([3.2, 2.8, 4.5, 5.0, 6.1])
x_pred = np.array([2.9, 3.0, 4.2, 4.8, 6.5])
```

```python
# 1
# MSE, MAE, RMSE, R-SQUARE using numpy
mse = np.mean((y_true - x_pred) ** 2)
mae = np.mean(np.abs(y_true - x_pred))
rmse = np.sqrt(mse)

ss_total = np.sum((y_true - np.mean(y_true)) ** 2)

ss_residual = np.sum((y_true - x_pred) ** 2)
r_square = 1 - (ss_residual / ss_total)
```

```python
print(f"MSE: {mse}, MAE: {mae}, RMSE: {rmse}, R-SQUARE: {r_square}")
```
