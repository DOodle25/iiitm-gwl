```python
import numpy as np
import pandas as pd

df = pd.read_csv("./archive/iris.csv")
```

```python
# 1
# convert to json and add new one
new_row_df = {
    "Id": df["Id"].max() + 1,
    "SepalLengthCm": 5.1,
    "SepalWidthCm": 3.5,
    "PetalLengthCm": 1.4,
    "PetalWidthCm": 0.2,
    "Species": "Iris-setosa"
}

df_new = pd.concat([df, pd.DataFrame([new_row_df])], ignore_index=True)
df_new.to_json("iris.json", orient="records", indent=2)
```

```python
# 2
# save new csv and add new
df.to_csv("iris_new.csv", index=False)

new_row_csv = {
    "Id": df["Id"].max() + 1,
    "SepalLengthCm": 5.1,
    "SepalWidthCm": 3.5,
    "PetalLengthCm": 1.4,
    "PetalWidthCm": 0.2,
    "Species": "Iris-setosa"
}

df_csv = pd.read_csv("iris_new.csv")
df_csv = pd.concat([df_csv, pd.DataFrame([new_row_csv])], ignore_index=True)
df_csv.to_csv("iris_new.csv", index=False)
```
