```python
 #! Q1. ImplemenƟng Simple Linear Regression from Scratch (OLS & GD)

import numpy as np
import pandas as pd

# a) Load the dataset and choose “area” as the feature and “price” as the target.
data = pd.read_csv("./archive/Housing.csv")
x = data['area'].values
y = data['price'].values
```

```python
# b) Write a function to implement Simple Linear Regression using Ordinary Least Squares (OLS) (derive slope & intercept).
import math
def ols_regression(x, y):
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    b1 = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - (np.sum(x))**2)
    b0 = y_mean - b1 * x_mean

    return b1, b0
b1, b0 = ols_regression(x, y)
print(b1, b0)

def ols_predict(x, b1, b0):
    return b0 + b1*x

print("predicted answer:", math.floor(ols_predict(7420, b1, b0)))
print("real answer:", 13300000)

def ols_predict(x, b1, b0):
    return b0 + b1 * x

y_pred = ols_predict(x, b1, b0)
errors = np.abs(y_pred - y)
median_error = np.median(errors)
print("Median error:", median_error)
```

```python
# c) Write another funcƟon for Simple Linear Regression using Gradient Descent (GD) with a learning rate and fixed iteraƟons.

def gd_regression(x, y, lr=1e-7, n_iter=1000):
    n = len(x)
    b0 = 0.0
    b1 = 0.0
    for _ in range(n_iter):
        y_pred = b0 + b1 * x
        error = y_pred - y
        b0 -= lr * (2/n) * np.sum(error)
        b1 -= lr * (2/n) * np.sum(error * x)
    return b1, b0

# b1_gd, b0_gd = gd_regression(x, y, lr=0.01, n_iter=10)
b1_gd, b0_gd = gd_regression(x, y, lr=1e-9, n_iter=10000)

print(b1_gd, b0_gd)

def gd_predict(x, b1, b0):
    return b0 + b1 * x

print("predicted answer:", math.floor(gd_predict(7420, b1_gd, b0_gd)))
print("real answer:", 13300000)

y_pred = gd_predict(x, b1_gd, b0_gd)
errors = np.abs(y_pred - y)
median_error = np.median(errors)
print("Median error:", median_error)
```

```python
# c) Implement Multiple Linear Regression using GD.

def gd_multi_regression(X_b, y, lr=1e-9, n_iter=10000):
    n, m = X_b.shape
    theta = np.zeros(m)
    for _ in range(n_iter):
        y_pred = X_b @ theta
        error = y_pred - y
        gradients = (2/n) * (X_b.T @ error)
        theta -= lr * gradients
    return theta

theta_gd = gd_multi_regression(X_b, y, lr=1e-9, n_iter=10000)
print("GD Coefficients:", theta_gd)

# def gd_multi_predict(X, theta):
#     # Add bias term to X if not already present
#     if X.shape[1] == theta.shape[0] - 1:
#         X_b = np.c_[np.ones((X.shape[0], 1)), X]
#     else:
#         X_b = X
#     return X_b @ theta

# # Example usage:
# y_pred_gd = gd_multi_predict(X, theta_gd)
# print("Predictions (first 5):", y_pred_gd[:5])
```
